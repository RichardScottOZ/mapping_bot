```python
import pandas as pd
import numpy as np
import geopandas as gpd
import pyproj
from shapely import geometry
import matplotlib.pyplot as plt
import earthpy

%matplotlib inline
```

```python
test = pd.read_csv('../../../database/csv/interpolado/cut_d_atest.csv')
```

```python
test['geometry'] = [geometry.Point(x, y) for x, y in zip(test['X'], test['Y'])]
crs = "+proj=utm +zone=21 +south +ellps=WGS84 +datum=WGS84 +units=m +no_defs"
test = gpd.GeoDataFrame(test, geometry='geometry', crs=crs)
test.head(100)
```

```python
test.plot(figsize=(15,15))
```

```python
test.to_csv('../../../database/csv/interpolado/')
```
